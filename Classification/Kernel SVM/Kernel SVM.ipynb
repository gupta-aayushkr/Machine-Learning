```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Social_Network_Ads.csv')
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25, random_state=0)
```

```python
print(X_train, len(X_train))
```

```python
print(X_test, len(X_test))
```

```python
print(y_train, len(y_train))
```

```python
print(y_test, len(y_test))
```

# Feature Scaling

In summary, use 
- sc.fit_transform on your training data to both learn scaling parameters and transform the data. 

Then, use 

- sc.transform on your test or validation data to apply the same scaling without relearning the parameters, ensuring consistency in the preprocessing. 

-This practice helps prevent data leakage and ensures that your model is trained and evaluated on data with the same preprocessing steps applied.

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

# Training Model

- only difference of Liner and Kernel where their Kernel paramenter in the mode

```python
from sklearn.svm import SVC
classifier = SVC(kernel='rbf', random_state=0) #metric and p value are used for ecludean distance
classifier.fit(X_train, y_train)
```

# Predit for Age 30 Salary 87k dollar
- feature scaling required

```python
print(classifier.predict(sc.transform([[40, 187000]])))
```

```python
y_pred = classifier.predict(X_test)
```

```python
y_pred = classifier.predict(X_test)
predict_result = np.concatenate((y_pred.reshape(len(X_test),1), y_test.reshape(len(y_test),1)),axis=1)
```

```python
print(predict_result)
```

# confusion Matrix

```python
from sklearn.metrics import confusion_matrix, accuracy_score
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```

```python
from matplotlib.colors import ListedColormap
X_set, y_set = sc.inverse_transform(X_test), y_test
X1, X2 = np.meshgrid(np.arange(start = X_set[:, 0].min() - 10, stop = X_set[:, 0].max() + 10, step = 1),
                     np.arange(start = X_set[:, 1].min() - 1000, stop = X_set[:, 1].max() + 1000, step = 1))
plt.contourf(X1, X2, classifier.predict(sc.transform(np.array([X1.ravel(), X2.ravel()]).T)).reshape(X1.shape),
             alpha = 0.75, cmap = ListedColormap(('red', 'green')))
plt.xlim(X1.min(), X1.max())
plt.ylim(X2.min(), X2.max())
for i, j in enumerate(np.unique(y_set)):
    plt.scatter(X_set[y_set == j, 0], X_set[y_set == j, 1], c = ListedColormap(('red', 'green'))(i), label = j)
plt.title('Kernvel SVM (Test set)')
plt.xlabel('Age')
plt.ylabel('Estimated Salary')
plt.legend()
plt.show()
```
