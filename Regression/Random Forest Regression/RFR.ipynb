```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Position_Salaries.csv')
X = dataset.iloc[:,1:-1].values
y = dataset.iloc[:,-1].values
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=10, random_state=0)
regressor.fit(X,y)
```

```python
regressor.predict([[6.5]])
```

```python
plt.scatter(X,y, color='red')
plt.plot(X, regressor.predict(X), color='blue')
plt.title('Random Forest Regression')
plt.xlabel('Experience')
plt.ylabel('Salary')
```

IN the Tutorial he has increased the data points on x axis therefore this prediction looks skewed according to me

```python
plt.scatter(X,y, color='red')
X_grid = np.arange(min(X), max(X), 0.1)  #creating more data points on X-axis
X_grid = X_grid.reshape(len(X_grid),1) #trandorming array in 2D matrix with only one col and many rows
plt.plot(X_grid, regressor.predict(X_grid), color='blue') #using X_grid for prediction
plt.title('Polinomial Regression')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.show()
```
